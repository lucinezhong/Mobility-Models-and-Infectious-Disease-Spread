```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_EPR=pd.read_csv('EPR_model_individual_trajectory.csv')
df_d_EPR=pd.read_csv('d_EPR_model_individual_trajectory.csv')
```

```python
df_EPR
```

```python
import matplotlib

def plot_trajecotry(df,title):
    fig, ax = plt.subplots(1, 1, figsize=(5, 4))
    
    
    ax.plot(df['from_lat'], df['from_lon'], marker='o', markersize=5,linestyle='-', color='b',zorder=1)

    ax.scatter(df['from_lat'][0:1], df['from_lon'][0:1], marker='*',color='red',s=30,zorder=2)
    # Add labels and title
    ax.set_xlabel('Longitude')
    ax.set_ylabel('Latitude')
    ax.set_title(title)
    
    # Add grid lines
    ax.grid(True)
    plt.tight_layout()
    #matplotlib.rcParams['savefig.dpi'] = 3000
    # Show the plot
    fig.savefig(title+'.eps')
```

```python
df=df_EPR[df_EPR['id']==2]
plot_trajecotry(df,'EPR model generated trajectory')
```

```python
df=df_d_EPR[df_d_EPR['id']==2]
plot_trajecotry(df,'d_EPR model generated trajectory')
```
